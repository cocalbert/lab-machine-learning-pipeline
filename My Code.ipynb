```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
```

```python
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
```

```python
#asigna un valor numÃ©rico a cada tipo de cut.
#asigna un valor numÃ©rico a cada tipo de clarity.
```

```python
cut_type = {
    'Ideal': 5,
    'Premium': 4,
    'Very Good': 3,
    'Good': 2,
    'Fair': 1
}

df_train['cut'] = df_train['cut'].map(cut_type)
```

```python
color_type = {
    'D': 7,
    'E': 6,
    'F': 5,
    'G': 4,
    'H': 3,
    'I': 2,
    'J': 1
}

df_train['color'] = df_train['color'].map(color_type)
```

```python
clarity_type = {
    'IF': 8,
    'VVS1': 7,
    'VVS2': 6,
    'VS1': 5,
    'VS2': 4,
    'SI1': 3,
    'SI2': 2,
    'I1': 1

}

df_train['clarity'] = df_train['clarity'].map(clarity_type)
```

```python
print(df_train.shape[0])
df_train.sample()
```

```python
print(df_test.shape[0])
df_test.sample()
```

```python
#se calcula la matriz de correlaciones entre todas las columnas del conjunto de datos del train.
```

```python
corr_mat = df_train.corr()
```

```python
df_train.drop(['depth','table'], axis = 1, inplace = True) 
```

```python
df_train
```

```python
#El conjunto de datos de entrenamiento se divide en variables independientes X (todas las columnas excepto "precio") y la variable dependiente y ("precio").
```

```python
X = df_train.iloc[:,:-1]
y = df_train['price']
```

```python
model = RandomForestRegressor(max_depth = 70, min_samples_leaf = 4, min_samples_split = 10, n_estimators = 400)
model.fit(X, y)
```

```python
#same transformations applied to the training dataset are applied to the corresponding columns in the test dataset (df_test). Then, the "depth" and "table" columns are dropped from the test dataset.
```

```python
df_test['cut'] = df_test['cut'].map(cut_type)
df_test['color'] = df_test['color'].map(color_type)
df_test['clarity'] = df_test['clarity'].map(clarity_type)
df_test.drop(['depth', 'table'], axis = 1, inplace = True) 
```

```python
# Los precios de los diamantes en este caso, se predicen utilizando el modelo entrenado y los datos de prueba.
```

```python
y_pred = model.predict(df_test)
y_pred
```

```python
# Los valores pronosticados se aÃ±aden como una nueva columna ("price") en el test dataset (df_test).
```

```python
df_test['price'] = y_pred
df_for_submission = df_test[["id", "price"]]
```

```python
df_for_submission.to_csv("my_submission.csv", index=False)
```
